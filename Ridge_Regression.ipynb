### Create a preprocessing pipeline for x

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
num_pipe =Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                         ('scaler', StandardScaler())])

cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                        ('ohe', OneHotEncoder(handle_unknown='ignore'))])

pre = ColumnTransformer([('num', num_pipe, con),
                        ('cat', cat_pipe, cat)])
```

```python
x_pre = pre.fit_transform(x).toarray()
x_pre
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
x_pre = pd.DataFrame(x_pre, columns=cols)
x_pre.head()
```

### Bulit a traintest model

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre, y, test_size=0.2, random_state=10)
```

### Tuning alpha value manually

```python
import numpy as np
alphas = np.arange(0.1,100,0.1)
alphas
```

```python
from sklearn.metrics import mean_squared_error
tr = []
ts = []
for i in alphas:
    # create and fit the model
    rr = Ridge(alpha=i)
    rr.fit(xtrain, ytrain)
    # Predict train and test
    ypred_tr = rr.predict(xtrain)
    ypred_ts = rr.predict(xtest)
    # Calculate train and test results
    tr_mse = mean_squared_error(ytrain, ypred_tr)
    ts_mse = mean_squared_error(ytest, ypred_ts)
    # append the result in tr and ts
    tr.append(tr_mse)
    ts.append(ts_mse)
```

```python
tr_mse
```

```python
ts_mse
```

```python
tr[0:5]
```

```python
ts[0:5]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(alphas, tr, label='Train MSE')
plt.plot(alphas, ts, label='Test MSE')
plt.xlabel('alpha')
plt.ylabel('MSE')
plt.title('Impact of alpha on train and test MSE')
plt.legend()
plt.show()
```

```python
for i in alphas:
    # create and fit the model
    rr = Ridge(alpha=i)
    rr.fit(xtrain, ytrain)
    # Predict train and test
    ypred_tr = rr.predict(xtrain)
    ypred_ts = rr.predict(xtest)
    # Calculate train and test results
    tr_mse = mean_squared_error(ytrain, ypred_tr)
    ts_mse = mean_squared_error(ytest, ypred_ts)
    # Check for optimal alpha
    if tr_mse >= ts_mse : 
        opt_alpha = i

        print(f'Optimal value of alpha is  = {opt_alpha}')
        break
```

```python
opt_alpha
```

### Create a final model with optimal alpha

```python
best_ridge = Ridge(alpha=opt_alpha)
best_ridge.fit(xtrain, ytrain)
```

```python
# R2 score in train
best_ridge.score(xtrain, ytrain)
```

```python
# R2 score in testing
best_ridge.score(xtest, ytest)
```

```python
evaluate_model(xtrain, ytrain, xtest, ytest, best_ridge)
```
