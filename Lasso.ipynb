### Seperate x and y variable

### Cat Con seperation for x variable

```python
cat = list(x.columns[x.dtypes=='object'])
con = list(x.columns[x.dtypes!='object'])
```

### Create a preprocessing pipeline for x

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                          ('scaler', StandardScaler())])

cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                          ('ohe', OneHotEncoder(handle_unknown='ignore'))])

pre = ColumnTransformer([('num', num_pipe, con),
                        ('cat', cat_pipe, cat)])
```

```python
x_pre = pre.fit_transform(x).toarray()
x_pre
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
x_pre = pd.DataFrame(x_pre, columns=cols)
x_pre.head()
```

### Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre, y, test_size=0.2, random_state=21)
```

### Manual Tuning of alpha

```python
import numpy as np
alphas = np.arange(0.1, 100, 0.1)
alphas
```

```python
from sklearn.metrics import mean_squared_error
tr = []
ts = []
for i in alphas:
    # create a lasso model
    ls = Lasso(alpha=i)
    ls.fit(xtrain, ytrain)
    # predict values for train and test
    ypred_tr = ls.predict(xtrain)
    ypres_ts = ls.predict(xtest)
    # Check MSE fro train and test
    tr_mse = mean_squared_error(ytrain, ypred_tr)
    ts_mse = mean_squared_error(ytest, ypres_ts)
    # Append in ts and tr
    tr.append(tr_mse)
    ts.append(ts_mse)
```

```python
tr[0:5]
```

```python
ts[0:5]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(alphas, tr, label='TR MSE')
plt.plot(alphas,ts, label='Ts MSE')
plt.xlabel('alpha')
plt.ylabel('MSE')
plt.title('Impact of Alpha on train and test result')
plt.legend()
plt.show()
```
